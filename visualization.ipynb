This notebook is intended for plotting image examples from the dataset

```python
import pandas as pd
import cv2
```

```python
train_df = pd.read_csv('train_fruits_dataset.csv')
test_df = pd.read_csv('test_fruits_dataset.csv')
```

```python
# Show random images from the training set
import matplotlib.pyplot as plt
import numpy as np

names = {
    "freshapples": 'Fresh Apples',
    "freshbanana": 'Fresh Banana',
    "freshoranges": 'Fresh Oranges',
    "rottenapples": 'Rotten Apples',
    "rottenbanana": 'Rotten Banana',
    "rottenoranges": 'Rotten Oranges'
}

fig, axes = plt.subplots(ncols=6, nrows=2, figsize=(10, 5))
axes = axes.ravel()
for i in range(12):
    index = np.random.randint(0, len(train_df))
    image = cv2.cvtColor(cv2.imread(train_df['image'][index]), cv2.COLOR_BGR2RGB)
    axes[i].imshow(image)
    axes[i].set_title(names[train_df['class'][index]])
    axes[i].get_xaxis().set_visible(False)
    axes[i].get_yaxis().set_visible(False)
plt.tight_layout()
plt.show()
```
